*** FUNCTION FOR CREATING AND RETURNING DATASET ***

```python
def crd(path):
  images=[]
  labels=[]
  c=0
  dict={}
  for i in os.listdir(path):
    c=c+1
    for j in os.listdir(path+r'/'+i):
      dict[c]=i
      img=cv2.imread(path+r'/'+i+r'/'+j)
      images.append(img)
      labels.append(c)
  return (np.asarray(images),np.asarray(labels),dict)
```

```python
images,labels,map = crd('/content/images/train')
```

***CONVERTING LABELS TO CATEGORICAL VALUE***

```python
labels=tf.keras.utils.to_categorical(labels)
```

***LOADING THE SAVED MODEL IF PRESENT***

```python
cnn=load_model('/content/drive/MyDrive/rtec/emotion1st.h5')
```

***CREATING THE MODEL***

```python
cnn=keras.models.Sequential([
                             keras.layers.Conv2D(50,kernel_size=(3,3),input_shape=(48,48,3)),
                             keras.layers.Conv2D(50,kernel_size=(3,3),activation='relu'),
                             keras.layers.MaxPool2D(2),
                             keras.layers.Dropout(0.3),
                             keras.layers.BatchNormalization(),
                             keras.layers.Conv2D(40,activation='relu',kernel_size=(3,3)),
                             keras.layers.Conv2D(40,activation='relu',kernel_size=(3,3)),
                             keras.layers.MaxPool2D(2),
                             keras.layers.Dropout(0.3),
                             keras.layers.BatchNormalization(),
                             keras.layers.Flatten(),
                             keras.layers.Dense(300,activation='relu'),
                             keras.layers.Dense(200,activation='relu'),
                             keras.layers.Dense(50,activation='relu'),
                             keras.layers.Dense(8,activation='softmax')
])
```

***HELPER FUNCTION TO EXTRACT FACE AND PASS TO THE MODEL FOR CLASSIFICATION***

```python
def predict(model,img_path,choice):
    if choice==1 :
         img = io.imread(img_path)
    elif choice==2 :
        img = cv2.imread(img_path)
    else:
        img=img_path
    plt.imshow(img)
    plt.show()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Load the cascade
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_alt2.xml')
    # Detect faces
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    for (y, x, w, h) in faces:
      img=img[x:x+w,y:y+h]
      plt.imshow(img)
      plt.show()
      break
    img=cv2.resize(img,(48,48))
    plt.imshow(img)
    plt.show()
    x=model.predict(np.asarray([img]))
    print(map[np.argmax(x)])
   
```

***SPLITTING THE DATASET TO TRAIN AND TEST***

```python
xtrain,xtest,ytrain,ytest=train_test_split(images,labels,train_size=0.7)
```

```python
xtrain.shape
```

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics='accuracy')
```

```python
history=cnn.fit(xtrain,ytrain,epochs=50,validation_split=0.2)
```

```python
plt.plot(history.history['accuracy'],c='blue')
plt.plot(history.history['loss'],c='red')
plt.xlabel('epochs')
plt.ylabel('loss vs accuracy')
plt.legend(['accuracy', 'loss'], loc='upper left')
plt.show()
```
