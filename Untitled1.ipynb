```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import cv2
import os
from PIL import Image
from xml.dom import minidom
from skimage import io
import zipfile
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
with zipfile.ZipFile('/content/drive/MyDrive/rtic/human face - Bing images.zip','r') as zp:
  zp.extractall('data')
```

```python

def crd(path):
  images=[]
  dim=[]
  c=0
  for i in sorted(os.listdir(path)):
    for j in sorted(os.listdir(path+r'/'+i)):
      if j.endswith('.xml'):
        obj=minidom.parse(path+r'/'+i+r'/'+j)
        xmin = int(obj.getElementsByTagName('xmin')[0].firstChild.nodeValue)
        ymin =int(obj.getElementsByTagName('ymin')[0].firstChild.nodeValue)
        xmax=int(obj.getElementsByTagName('xmax')[0].firstChild.nodeValue)
        ymax=int(obj.getElementsByTagName('ymax')[0].firstChild.nodeValue)
        dim.append([xmin,ymin,xmax,ymax])
      else:
        img=cv2.imread(path+r'/'+i+r'/'+j)
        images.append(img)
    c=c+1

  return (np.asarray(images),np.asarray(dim))
```

```python
images,dim=crd("/content/data")
```

```python
dim.shape
```

```python
bdb=keras.models.Sequential([
                             keras.layers.Input(shape=(None,None,3)),
                             keras.layers.Resizing(220,220),
                             keras.layers.Conv2D(100,kernel_size=(3,3),activation='relu'),
                             keras.layers.Conv2D(120,kernel_size=(3,3),activation='relu'),
                             keras.layers.MaxPool2D(2),
                             keras.layers.Conv2D(100,activation='relu',kernel_size=(3,3)),
                             keras.layers.Conv2D(90,activation='relu',kernel_size=(3,3)),
                             keras.layers.MaxPool2D(2),
                             keras.layers.Conv2D(50,activation='relu',kernel_size=(3,3)),
                             keras.layers.MaxPool2D(2),
                             keras.layers.Flatten(),
                             keras.layers.Dense(400,activation='relu'),
                             keras.layers.Dense(200,activation='relu'),
                             keras.layers.Dense(100,activation='relu'),
                             keras.layers.Dense(30,activation='relu'),
                             keras.layers.Dense(4,activation='relu'),      
])
```

```python
bdb.compile(optimizer='adam',loss=['huber'])
```

```python
bdb.fit(images,dim,epochs=40)
```
